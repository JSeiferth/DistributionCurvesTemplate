```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters for the linear distribution
slope = -0.01  # The slope of the line (negative to go from top left to bottom right)
intercept = 1  # The intercept of the line (set to a value to start from the top left)

# Generate x values
x = np.linspace(0, 100, 100)

# Compute the corresponding y values using the linear equation y = mx + c
y = slope * x + intercept

# Plot the linear distribution curve
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=f'y = {slope}x + {intercept}', color='red')
plt.title('Linear Distribution Curve')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters for the power-law distribution
alpha = 0.4  # The exponent for the power-law distribution (lower value for a less steep curve)
C = 1000000  # The scaling constant

# Generate x values
x = np.linspace(1, 100, 100)  # Avoid x=0 since power-law is undefined at x=0

# Compute the corresponding y values using the power-law equation y = C * x^(-alpha)
y = C * x**(-alpha)

# Plot the power-law distribution curve
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$y = C \cdot x^{-\alpha}$', color='red')
plt.title('Power-Law Distribution Curve (Linear Scale)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True, linestyle='--', linewidth=0.5)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters for the step function
x = np.array([1, 2, 3, 4, 5])  # Step positions on the x-axis
y = np.array([325000, 245000, 165000, 85000, 8000])  # Corresponding y values

# Plot the step function
plt.figure(figsize=(6, 6))
plt.step(x, y, where='post', color='red', linewidth=2)  # where='post' aligns the steps to the right
plt.xticks([])  # Remove x-axis ticks
plt.yticks([8000, 85000, 165000, 245000, 325000], ['8K', '85K', '165K', '245K', '325K'])  # Custom y-axis ticks
plt.title('Impact groups', fontsize=14, fontweight='bold')
plt.gca().set_facecolor('#f5f5f5')  # Set background color to light grey
plt.grid(True, which="both", linestyle='--', linewidth=0.5)

# Customize the plot to resemble the image closely
plt.box(False)  # Remove the plot box
plt.gca().spines['left'].set_visible(False)  # Hide the left spine
plt.gca().spines['bottom'].set_visible(False)  # Hide the bottom spine
plt.gca().spines['right'].set_visible(False)  # Hide the right spine
plt.gca().spines['top'].set_visible(False)  # Hide the top spine

plt.show()


```
